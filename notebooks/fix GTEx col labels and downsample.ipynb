## Fix column label tuple format
It seems that having parentheses causes problems in parsing the tuple, so I'm going to rename ``(Detailed)`` to ``-Detailed``.

```python
cols = gtex.columns.tolist()

new_cols = []
for inst_col in cols:
    inst_col = make_tuple(inst_col.strip().replace('(Detailed)','-Detailed'))
    new_cols.append(inst_col)

gtex.columns = new_cols
```

```python
gtex.to_csv('../hzome_data/my_gtex_Moshe_2017_exp_clean_cols.txt', sep='\t')
```
