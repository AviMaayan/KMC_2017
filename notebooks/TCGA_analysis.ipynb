# Investigating PKMYT1 expression across TCGA tumor types

## Overview
In order to investigate the role of ‘dark’ genes in cancer, we analyzed their expression and their association to patient survival across thousands of tumor samples profiled by The Cancer Genome Atlas (TCGA) project.

This notebook focuses on the analysis a ‘dark’ kinase which is part of a larger cluster of kinases involved in the regulation of cell cycle: PKMYT1.  These genes were identified by performing coexpression analysis on gene expression data of cancer cell lines from CCLE (link to the notebook here).

### Differential Expression Analysis
PKMYT1 is a member of the serine/threonine protein kinase family and is known to be involved in the regulation of the G2/M transition of the cell cycle.


```python
Image('../img/TCGA_notebook_images/pkmyt_differential_expression.png')
```

**Figure 1 | PKMYT1 is significantly overexpressed in multiple TCGA tumor types.**  The boxplots display the expression of PKMYT1, measured in log10FPKM, in tumor samples and matched normal tissue.  The kinase is significantly overexpressed in a variety of carcinomas, including Breast Carcinoma, Lung Squamous Carcinoma, and Kidney Renal Clear Cell Carcinoma.

### Survival Analysis

```python
Image('../img/TCGA_notebook_images/pkmyt_survival.png')
```

**Figure 2 | PKMYT1 is associated to poor survival in patients of Kidney Renal Clear Cell Carcinoma (KIRC) and Adrenocortical Carcinoma (ACC).**  Patients from TCGA were divided in three groups, according to the relative expression of PKMYT1 (high, medium, low).  These groups have significantly different survival life, as shown in the Kaplan-Meier curves.  P-values were calculated using an implementation of the log-rank test in the survival R package.
